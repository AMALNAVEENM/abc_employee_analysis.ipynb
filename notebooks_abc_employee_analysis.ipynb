```python
# ----------------------------
# 1. IMPORT LIBRARIES
# ----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8')
```

```python
# ----------------------------
# 2. LOAD AND PREPROCESS DATA
# ----------------------------
# Load dataset (replace 'abc_company.csv' with your file)
df = pd.read_csv('abc_company.csv')

# Fix height column: replace with random values between 150–180
np.random.seed(42)  # for reproducibility
df['height'] = np.random.randint(150, 181, size=len(df))

print("✅ Data shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Calculate team distribution
team_counts = df['team'].value_counts()
team_pct = (team_counts / len(df)) * 100

# Visualization
plt.figure(figsize=(10, 6))
ax = team_counts.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)

# Add percentage labels
for i, v in enumerate(team_counts):
    ax.text(i, v + 2, f"{team_pct.iloc[i]:.1f}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("Team Distribution (%):")
print(team_pct.round(2))
```

```python
# Group by position
position_counts = df['position'].value_counts()

# Visualization
plt.figure(figsize=(12, 6))
position_counts.plot(kind='barh', color='lightcoral')
plt.title('Employee Count by Position')
plt.xlabel('Number of Employees')
plt.ylabel('Position')
plt.tight_layout()
plt.show()

print("Top 5 Positions:")
print(position_counts.head())
```

```python
# Create age groups
bins = [20, 30, 40, 50, 60]
labels = ['20-29', '30-39', '40-49', '50-59']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

age_group_counts = df['age_group'].value_counts().sort_index()

# Visualization
plt.figure(figsize=(8, 5))
age_group_counts.plot(kind='bar', color='mediumseagreen')
plt.title('Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print("Predominant Age Group:", age_group_counts.idxmax())
```

```python
# Total salary by team and position
team_salary = df.groupby('team')['salary'].sum().sort_values(ascending=False)
pos_salary = df.groupby('position')['salary'].sum().sort_values(ascending=False)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Team
team_salary.head().plot(kind='bar', ax=axes[0], color='gold')
axes[0].set_title('Top Teams by Total Salary')
axes[0].set_ylabel('Total Salary ($)')

# Position
pos_salary.head().plot(kind='bar', ax=axes[1], color='purple')
axes[1].set_title('Top Positions by Total Salary')
axes[1].set_ylabel('Total Salary ($)')

plt.tight_layout()
plt.show()

print("Team with highest salary expenditure:", team_salary.index[0])
print("Position with highest salary expenditure:", pos_salary.index[0])
```

```python
# Calculate correlation
corr = df['age'].corr(df['salary'])
print(f"Correlation between Age and Salary: {corr:.3f}")

# Visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='salary', alpha=0.7)
sns.regplot(data=df, x='age', y='salary', scatter=False, color='red')
plt.title(f'Age vs Salary (r = {corr:.3f})')
plt.xlabel('Age')
plt.ylabel('Salary ($)')
plt.tight_layout()
plt.show()
```
